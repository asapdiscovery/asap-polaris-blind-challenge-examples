```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
from glob import glob
import matplotlib.patches as patches
from matplotlib.patches import FancyBboxPatch
```

```python
import polaris as po

# Load the datasets from the Hub
dataset = po.load_dataset("asap-discovery/antiviral-admet-2025-unblinded")
df_admet = pd.DataFrame(dataset[:])

dataset = po.load_dataset("asap-discovery/antiviral-potency-2025-unblinded")
df_potency = pd.DataFrame(dataset[:])

dataset = po.load_dataset("asap-discovery/antiviral-ligand-poses-2025-unblinded")
df_poses = pd.DataFrame(dataset[:])
df_poses
```

```python
df_admet
```

```python
poses_train = df_poses[df_poses["Set"] == "Train"]
poses_test = df_poses[df_poses["Set"] == "Test"]
poses_test_sars2 = poses_test[poses_test["Protein Label"] == "SARS-CoV-2 Mpro"]
poses_test_mers = poses_test[poses_test["Protein Label"] == "MERS-CoV Mpro"]

mers_cov_potency_train = df_potency[df_potency["Set"] == "Train"]["pIC50 (MERS-CoV Mpro)"].dropna()
sars_cov_2_potency_train = df_potency[df_potency["Set"] == "Train"]["pIC50 (SARS-CoV-2 Mpro)"].dropna()

admet_hlm_train = df_admet[df_admet["Set"] == "Train"]["HLM"].dropna()
admet_mlm_train = df_admet[df_admet["Set"] == "Train"]["MLM"].dropna()
admet_ksol_train = df_admet[df_admet["Set"] == "Train"]["KSOL"].dropna()
admet_mdck2_train = df_admet[df_admet["Set"] == "Train"]["MDR1-MDCKII"].dropna()
admet_logd_train = df_admet[df_admet["Set"] == "Train"]["LogD"].dropna()

##

mers_cov_potency_test = df_potency[df_potency["Set"] == "Test"]["pIC50 (MERS-CoV Mpro)"].dropna()
sars_cov_2_potency_test = df_potency[df_potency["Set"] == "Test"]["pIC50 (SARS-CoV-2 Mpro)"].dropna()

admet_hlm_test = df_admet[df_admet["Set"] == "Test"]["HLM"].dropna()
admet_mlm_test = df_admet[df_admet["Set"] == "Test"]["MLM"].dropna()
admet_ksol_test = df_admet[df_admet["Set"] == "Test"]["KSOL"].dropna()
admet_mdck2_test = df_admet[df_admet["Set"] == "Test"]["MDR1-MDCKII"].dropna()
admet_logd_test = df_admet[df_admet["Set"] == "Test"]["LogD"].dropna()
```

```python
subchallenges = (
                 "Ligand posing", 
                 "Potency (MERS-CoV)", "Potency (SARS-CoV-2)", 
                 "ADMET:HLM", "ADMET:MLM", "ADMET:KSOL", "ADMET:MDR1-MDCKII", "ADMET:LogD",
                 )
quantities = {
    'Training sets': (
        len(poses_train),
        len(mers_cov_potency_train),
        len(sars_cov_2_potency_train),
        len(admet_hlm_train),
        len(admet_mlm_train),
        len(admet_ksol_train),
        len(admet_mdck2_train),
        len(admet_logd_train),
        ),
    'Test sets': (
        len(poses_test_sars2),
        len(mers_cov_potency_test),
        len(sars_cov_2_potency_test),
        len(admet_hlm_test),
        len(admet_mlm_test),
        len(admet_ksol_test),
        len(admet_mdck2_test),
        len(admet_logd_test),
        ),
}

x = np.arange(len(subchallenges))  # the label locations
width = 0.4  # the width of the bars
multiplier = 0.5

fig, ax = plt.subplots(layout='constrained', figsize=(6,6))

for attribute, measurement in quantities.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute, edgecolor="black")
    ax.bar_label(rects, padding=2, size=8)
    multiplier += 1

# stack the MERS poses bar on top
rects = ax.bar(0 + offset, len(poses_test_sars2)+len(poses_test_mers), width, color="tab:orange", edgecolor="black", zorder=-1)
print(rects)
ax.text(0+offset, y=len(poses_test_sars2)+len(poses_test_mers)+8,s=str(len(poses_test_mers)), ha="center", size=8)
ax.annotate("M", xy=(0.5, 150), color="white")

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.annotate("COVID Moonshot: SARS-CoV-2 Mpro", xy=(0.12, 25), color="white", rotation=90)
ax.annotate("S", xy=(0.5, 50), color="white")
ax.set_ylabel('Number of compound measurements in set')
# ax.set_title('ASAP:Polaris blind challenges for SARS-CoV-2/MERS-CoV Mpro')
ax.set_xticks(x + width, subchallenges, rotation=30, ha="right")
ax.legend(loc=(0.4, 0.6), ncols=3)
ax.set_ylim(0, 940)



# style = "Simple, tail_width=0.5, head_width=4, head_length=8"

# a1 = patches.FancyArrowPatch((4, 830), (2.4,700),
#                              connectionstyle="arc3,rad=-.5", **dict(arrowstyle=style, color="tab:blue"))
# a2 = patches.FancyArrowPatch((4, 600), (2.6,300),
#                              connectionstyle="arc3,rad=.5", **dict(arrowstyle=style, color="tab:orange"))
# for a in [a1, a2]:
#     plt.gca().add_patch(a)

# ax.annotate(r"$\mathbf{Training\ sets}$"+" that participants will\nfinetune their models with:\n"+r"$\mathit{these\ contain\ experimental\ data.}$", xy=(0.52, 0.8), xycoords="axes fraction")

# ax.annotate(r"$\mathbf{Test\ sets}$"+" that participants will\npredict on with their models:\n"+r"$\mathit{these\ only\ contain\ SMILES\ data.}$", xy=(0.52, 0.5), xycoords="axes fraction")

plt.savefig(f"datasets_overview.png", dpi=300)
plt.show()
```
